```python
import yfinance as yf
import pandas as pd
from datetime import timedelta, date

tsla = yf.Ticker("TSLA")
hist = tsla.history(period='max')
```

```python
tesla_earnings_dates = ['2021-01-27', '2020-10-21', '2020-07-22', '2020-04-29', '2020-01-29', '2019-10-23', '2019-06-24', '2019-04-24', '2019-01-30', '2018-10-24', '2018-08-01', '2018-05-02', '2018-02-07', '2017-11-01', '2017-08-02', '2017-05-03', '2017-02-22', '2016-10-26', '2016-08-03', '2016-05-04', '2016-02-10', '2015-11-03', '2015-08-05', '2015-05-06', '2015-02-11', '2014-11-05', '2014-07-31', '2014-05-07', '2014-02-19', '2013-11-05', '2013-08-07', '2013-05-08', '2013-02-20', '2012-11-05', '2012-07-25', '2012-05-08', '2012-02-15', '2011-11-01', '2011-08-03']

closes_before_earnings = hist.loc[tesla_earnings_dates, ['Open', 'Close']]
closes_before_earnings = closes_before_earnings.assign(weekday=lambda df: pd.to_datetime(df.index).weekday + 1)
closes_before_earnings = closes_before_earnings.assign(next_day=lambda df: pd.to_datetime(df.index) + timedelta(days=1))
closes_before_earnings['next_day_open'] = closes_before_earnings.apply(lambda r: hist.loc[r.next_day, 'Open'], axis=1)
closes_before_earnings['next_day_close'] = closes_before_earnings.apply(lambda r: hist.loc[r.next_day, 'Close'], axis=1)
closes_before_earnings['diff_closes'] = closes_before_earnings.apply(lambda r: r.next_day_close - r.Close, axis=1)
closes_before_earnings['diff_closes_%'] = closes_before_earnings.apply(lambda r: 100*(1 - (r.Close / r.next_day_close)), axis=1)
closes_before_earnings['diff_opens'] = closes_before_earnings.apply(lambda r: r.next_day_open - r.Open, axis=1)
closes_before_earnings['diff_opens_%'] = closes_before_earnings.apply(lambda r: 100*(1 - (r.Open / r.next_day_open)), axis=1)
closes_before_earnings['diff_close2open'] = closes_before_earnings.apply(lambda r: r.next_day_open - r.Close, axis=1)
closes_before_earnings['diff_close2open_%'] = closes_before_earnings.apply(lambda r: 100*(1 - (r.Close / r.next_day_close)), axis=1)
closes_before_earnings
```
